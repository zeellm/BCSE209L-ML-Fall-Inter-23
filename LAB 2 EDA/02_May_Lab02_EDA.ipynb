
# Checking the missing values


```python
df_market.isnull().sum()
```

#Drop the missing Values


```python
# Dropping the records with age missing in data dataframe.
df_market = df_market[~df_market.age.isnull()].copy()

# Checking the missing values in the dataset.
df_market.isnull().sum()
```

#Impute values to the missing values

```python
# Find the mode of month in data
month_mode = df_market.month.mode()[0]
print(month_mode)
# Fill the missing values with mode value of month in data.
df_market.month.fillna(month_mode, inplace = True)

# Let's see the null values in the month column.
df_market.month.isnull().sum()
```

```python
#drop the records with response missing in data.
df_market = df_market[~df_market.response.isnull()].copy()
# Calculate the missing values in each column of data frame
df_market.isnull().sum()
```

#Handling Outliers


**Univariate outliers**: Univariate outliers are the data points whose values lie beyond the range of expected values based on one variable.

**Multivariate outliers:** While plotting data, some values of one variable may not lie beyond the expected range, but when you plot the data with some other variable, these values may lie far from the expected value.

Categorical Unordered Univariate Analysis

```python
# Let's calculate the percentage of each job status category.
df_market.job.value_counts()

#plot the bar graph of percentage job categories
df_market.job.value_counts().plot.bar()
plt.show()
```

Categorical Ordered Univariate Analysis:


```python
#calculate the percentage of each education category.
df_market.education.value_counts(normalize=True)

#plot the pie chart of education categories
df_market.education.value_counts(normalize=True).plot.pie()
plt.show()
```

```python
df_market.salary.describe()
```

Bivariate Analysis - *Numeric*-Numeric Analysis:


```python
#plot the scatter plot of balance and salary variable in data
plt.scatter(df_market.salary,df_market.balance)
plt.show()

#plot the scatter plot of balance and age variable in data
df_market.plot.scatter(x="age",y="balance")
plt.show()
```

```python
#plot the pair plot of salary, balance and age in data dataframe.
sns.pairplot(data = df_market, vars=['salary','balance','age'])
plt.show()
```

Correlation Matrix

```python
# Creating a matrix using age, salry, balance as rows and columns
df_market[['age','salary','balance']].corr()

#plot the correlation matrix of salary, balance and age in data dataframe.
sns.heatmap(df_market[['age','salary','balance']].corr(), annot=True, cmap = 'Reds')
plt.show()
```

Bivariate Analysis - Numeric - Categorical Analysis



```python
#groupby the response to find the mean of the salary with response no & yes separately.
df_market.groupby('response')['salary'].mean()
```

```python
df_market.groupby('response')['salary'].median()
```

```python
sns.boxplot(x=df_market.response,y= df_market.salary)
plt.show()
```

Categorical — Categorical Analysis

```python
df_market['response_rate'] = np.where(df_market.response=='yes',1,0)
df_market.response_rate.value_counts()
```

```python
#plot the bar graph of marital status with average value of response_rate
df_market.groupby('marital')['response_rate'].mean().plot.bar()
plt.show()
```

Multivariate Analysis


```python
result = pd.pivot_table(data=df_market, index='education', columns='marital',values='response_rate')
print(result)
```

```python
#create heat map of education vs marital vs response_rate
sns.heatmap(result, annot=True, cmap = 'RdYlGn', center=0.117)
plt.show()
```
